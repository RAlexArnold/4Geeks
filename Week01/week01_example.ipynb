**DataFrames** are one of two main ways data is stored in **pandas**, the other being a **Series**

These ways of storing, or structuring data are called **Data structures**. So **DataFrames** and **Series** are types of **Data Structures**.

DataFrames are like spreadsheets. They are two-dimensional data structures.

They are like 2D arrays in **Numpy**, but with some differences.

1.) DataFrames have row and column indices
    
   - **Row Index** An array of labels associated to the rows
    
   - **Column Index** An array of labels associated to the columns
   
2.) Each column in a DataFrame can store a different **Data Type** (how data is stored, e.g. **float**, **integer**, **string**, **object**, etc.), while for Numpy all values in the array will have the same data type.
    

Let's focus on point 1.)

Imagine a dataframe presented below.

- The **Data** is the array of numerica values, (*[3, 0], [2, 3], ....* )

- The **Row Index** is the label of each row (*0, 1, 2, 3*). These are commonly integers but they don't have to be!

- The **Column Index** is the label of each columns (*apples, oranges*)

**Accessing Values, Rows, Columns**

```python
car_df
```

With Pandas DataFrames, we can access data by a rows and column labels using `.loc`

`.loc` requires the actual row/column labels to be used

```python
car_df.loc[3, 'Model'] # <- So this works
```

```python
car_df.loc[3, 1] # <- But what about this!??
```

`loc` requires the original/actual index labels. There is no column label called "1" so the above fails

Let's continue with `loc`

```python
# We can select a single row...
car_df.loc[2] #< - Notice the output of this looks a little different. 
# It's a Series, not a DataFrame. We'll come back to this in a moment
```

```python
# ... or multiple rows (place the row indices between square brackets)
car_df.loc[[2, 5]] 
```

When using doulbe brackets with `loc`, the *first set* of inner-most brackets signify the row-indices we use. So in the above, we will use rows 2 and 5.


Any *second set* of innter-most brackets will signify the column labels we use. 

So in the below we will use rows 3 and 5 (the first set of inner most brackets),

and we will use columns `Model` and `Color` (the second set of inner most brackets).

```python
# We can select multiple rows and multiple columns
car_df.loc[[3,5], ['Model', 'Color']]
```

Notice that using double brackets will return a **DataFrame**, while a single bracket will return a **Series**, even if we use a single row. 

We'll explore what a **Series** in a bit.

But for now to get a taste of **DataFrames** vs **Series**, compare...

```python
car_df.loc[3]
```

... to

```python
car_df.loc[[3]]
```

Again, notice how using double brackets changes the output.

```python
# A series...
car_df.loc[3, 'Model']
```

```python
# vs. a DataFrame
car_df.loc[[3],['Model']]
```

We can also access data by rows and columns using `.iloc`. 

But `.iloc` requires the *integer position* of each row and each column.

It doesn't want the original index label. It wants an integer.

```python
# So check what happens here. 
car_df.iloc[3, 'Model'] # <- ! 'Model' is not an integer position
```

```python
car_df.iloc[3, 1] # So if we are using .iloc we need the integer position of the row label and the column label
```

```python
# We can also grab multiple rows and columns with .loc
car_df.loc[[5,6,7], ['Model', 'Color', 'Value1']]
```

```python
# Or we could use .iloc.
car_df.iloc[[5,6,7], [2,3,4]]
```

```python
# Showing the car_df again 
car_df
```

```python
# There are also a few ways to grab all values of a column or columns
# The ":" in the row index position means to use all rows

car_df.loc[:, 'Color'] #<- Notice the output looks different
```

```python
# Or just
car_df['Color']
```

Notice that accessing values form DataFrames differs from how we access values from Numpy arrays

Numpy arrays have no index, so we just use the integer position (row, column) of the value of interest

```python
car_array = np.array(car_df)
car_array
```

```python
# Accessing a value in an array
car_array[1,2] 
```

Let's return to our `car_df` dataframe

```python
# Let's grab the Color column again
car_df['Color']
```

You may have noticed the output of the above looks slightly different than our DataFrame outputs. It doesn't appear to have a **column index**, only **row indices**

That's because the above is a **Series**, not a dataframe.

A **Series** is a one-dimensional labeled data structure. It has a **row index**, but lacks a **column index**

```python
car_df # Show the dataframe for car_df
```

```python
color_series = car_df['Color'] #Selecting a column with single brackets returns a series of that column
color_series
```

We can still access values using `loc` or `.iloc`

```python
color_series.loc[3]
```

But note, that if we use only a bracket (no `loc` or `iloc`) on a series, it assumes a row index is given in the bracket

```python
color_series[3]
```

If we wanted to keep the `Color` information from `car_df` as a **dataframe**, and not a **series** then we need to use double brackets

```python
car_df[['Color']]
```

# Filtering

We can also filter dataframes using a combinations of **logical conditions** and `.loc` or `.iloc`

Let's return to our Pokemon dataframe, `df`.

```python
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.columns
```

Suppose we want to grab only the Pokemon (rows) whose primary types are Grass

```python
df['Type 1']
```

```python
# Write the logical condition. 
# This returns a series boolean values
df['Type 1'] == 'Grass'
```

If we have a **series** of **booleans** (`True` or `False`) then we can filter the rows of our dataframe with this boolean series.

But for this to work, our boolean series must be the same length as the number of rows in our dataframe

Here, we made the boolean series using our Pokemon dataframe, so the lengths will match.

This boolean series used for filtering is called a **mask**

We can filter our dataframe using the **mask** in two ways.

Note, that this will keep rows where the **mask** is `True`, and remove rows where the **mask** is `False`

```python
mask = df['Type 1'] == 'Grass'

# Method 1 with loc
df.loc[mask]
```

```python
# Method 2
df[mask]
```

```python
# We could have also just written the above as
df[df['Type 1'] == 'Grass']
```

# Handling NaNs

Often you'll want to check if there are any NaN (Not a Number) values in your dataset. 

These are like missing data, and can often "throw a wrench" in your calculations

Let's return to our Pokemon Data to check if 

```python
na_data = [[1, 2, np.nan, np.nan],
           [3, np.nan, np.nan, np.nan],
           [2, np.nan, np.nan, np.nan],
           [3, 3, 4, np.nan],
           [1, 3, 2, np.nan]]


nan_df = pd.DataFrame(na_data)  

nan_df
```

There are many ways to drop NaN values depending your requirements

The `dropna` function is used to remove nan values

See docs: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html

By default, it drops **rows** from the dataframe containing any (at least one) NaN

```python
nan_df.dropna()
```

We can change this to drop **columns** with any (at least one) NaN by changing the **axis** keyword to axis=1

```python
nan_df.dropna(axis=1)
```

We can also make a change so that it drops **rows** containing *all* NaNs 

```python
nan_df.dropna(how='all') #<- Here, there are no rows that contain ALL NaNs, so no change
```

And we can make a change to drop **columns** containing *all* NaNs

```python
nan_df
```

```python
nan_df.dropna(axis=0, how='all')
```
