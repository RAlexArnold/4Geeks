For *any kind* of plot, you can change the size, add labels, change colors, add a title, change the axis scale, and more...

- Change the figure size by first adding `plt.figure(figsize=(width, height)`
- Add labels by adding the `label='text'` parameter *inside* your plotting function
- Change colors by adding the `c='color'` parameter *inside* your plotting function
- Change the scale by adding a new line that states `plt.xscale('log')` or `plt.yscale('log')`
- Add axis labels by adding a new line that states `plt.xlabel('text')` or `plt.ylabel('text')`
- You can change the transparency of the visualization by adding the `alpha=float` parameter  parameter *inside* your plotting function
    - `alpha` must be between 0 and 1. 
    - 0 is absolutely transparent and 1 is absolutely opaque
- Add a title by adding a new line that states `plt.title('text')`

See docs for colors you can use: https://matplotlib.org/stable/gallery/color/named_colors.html

`matplotlib.pyplot` will automatically cycle through the colors if there are multiple lines being plotted.

- The first (auto) color is called `'C0'`, then `'C1'`, ..., up to `C9`
- You can also specified a named color such as
    - `'b'` for blue
    - `'r'` for red
    - `'k'` for black, ...

# Line Plots

See docs for the `plot` function: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
# Just some data to play with
x = np.linspace(0, 10, 100)
y = np.exp(x)

# start with plt.plot(x,y)
# then add a label. show it does nothing. Must add plt.legend
# Then plot 10*y
# Then plot np.exp(2*x)
# Then show yscale('log')
# Then show colors
# You can show alpha too
# Show a title

plt.plot(x,y, c='blue')
plt.plot(x,10*y, c='r')
plt.plot(x, np.exp(2*x) , c='k', alpha=0.1)

plt.yscale('log')

plt.xlabel('x')
plt.ylabel('exp(x)')
plt.title('Exponential')

plt.show()
```

For a lineplot (`plot`) you can also change the **linestyle** (`ls`) and the **linewidth** (`lw`)

# Histograms

If we want to focus on the distribution of a single feature (variable) then we can use a histogram

We can set the bins using the `bins` parameter in the `plt.hist` function.

- Can be an integer specifying number of bins, i.e. `bins=10` (default is 10)
- Can be a list or array specifiying the ends points of the bins
    - e.g. `bins=[1,2,3,4]` would be 3 bins, each bin is between the specified points
    - use `np.arange(start, stop, step)` to generate array from `start` to `stop-step` (last point not included) with spacing of `step`
    - use `np.linspace(start, stop, num)` to generate array from `start` to `stop` (last point included) with `num` elements
    - use `np.logspace(start, stop, num)` to generate array like in `linspace`, but values are even spaced in a **log scale**
        - Beginning value of log space is `base ** start` and ending value is `base ** end`
        - So `start=2` corresonds to a starting value of 100 if the `base` keyword is `base=10`.
        
        
We can change the visual style of the histogram using the `histtype` parameter
- `histtype=bar` (default) Traditional style
- `histtype=step` Unfilled bars

Many other parameters. See docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

```python
np.arange(0, 100+5, 5)

np.linspace(0, 300, 50)
```

# Bar Plots and Pie Charts for Categorical Data

When we have **categorical data**, that is data that can easily be divided or represented by groups, then a **bar plot** or **pie chart** is best.

```python
df
```

Let's use the `quality` feature to make categorical data. Use the `quality` as a category.

Let's count how many data points there are for each quality-category

```python
quality_counts = df['quality'].value_counts()
quality_counts
```

Let's make a bar plot of the counts for each quality group.

If we tried this with a **histogram**, check what happens

```python
plt.hist(quality_counts)
```

Doesn't give us what we want. So we must use a **bar plot** using `plt.bar(x, height)`

```python
plt.bar(quality_counts) #<--- this alone won't work. Need an 'x' and 'y' separately
```

```python
print(quality_counts.index) #<--- these will be our category labels
print()
print(quality_counts.values) #<--- these will be the counts that are plotted for each category. 
# Series.values turns a series into a numpy array
```

```python
plt.bar(quality_counts.index, quality_counts)
plt.xlabel('Quality Value')
plt.ylabel('Number of Wines')
```

Bar plots don't just have to plot counts, we could use `groupby` to find various values of each quality-category

Below, we find the mean pH of each quality group

```python
df
```

```python
quality_pH = df.groupby(['quality'])['pH'].mean()
quality_pH
```

```python
plt.bar(quality_pH.index, quality_pH)
plt.xlabel('Quality Value')
plt.ylabel('pH')
```

It's hard to tell the bars apart, so we can change the y-axis limits using `plt.ylim(min, max)` (`plt.xlim` also exists)

Note that `plt.ylim` works for *any kind* of plot, not just bar plots

```python
plt.bar(quality_pH.index, quality_pH)
plt.xlabel('Quality Value')
plt.ylabel('pH')
plt.ylim(3.0,3.5)
```

Let's go back to our counts for each quality and plot these on a **pie chart**

```python
quality_counts = df['quality'].value_counts()
quality_counts
```

```python
plt.pie(quality_counts, labels=quality_counts.index)
plt.title('Pie Chart of Quality Counts')
plt.show()
```

# Subplots

Notice that each of the above plot returned an `<Axes: >` object. 

In all our previous examples, matplotlib was making what are called axes objects.

An axes object is like a blank canvas that can hold and draw your plots. It is a space where you can place your data opints, lines, bars, etc.

We can have more flexibility and customizing with how we plot if we start working with these axes objects directly!

But it can get a big more challenging. So once you are comfortable with the above approaches to plotting, you can start modifying axes objects.

This is where **subplots** come into play.

`f, axs = plt.subplots(n_rows, n_cols)`

```python
f, axs = plt.subplots(6, 4, figsize=(5,10))   
```

```python
axs[1,3].hist(df['total sulfur dioxide'])
f
```

```python
axs[0].hist(df['total sulfur dioxide']) #<---- but where is the plot?
```

```python
f      #<--- there it is 

# f is the variable that stores the final figure image.

# It'll be eaiser it we keep our plotting all in one cell.
```

Let's keep our (sub-)plotting in one cell to avoid having to retype `f`, the figure variable

Some of our commands for modifying our plots haved changed now for subplots

e.g.

- use `.set_xlabel` instead of `xlabel` to change each axes-object's axis-label titles
- use `.set_title` instead of `title` to change each axes-object's title
- use `.set_xlim` instead of `xlim` to change each axes-oject's axis-limits

- ... typically, for `subplots` you use `.set_` before the name of the function to change something

- use `.suptitle` to set an overall title for the entire figure

- You can set the parameters `sharex` and `sharey` to a boolean values in the `subplots` function to share axis values between axes-objects

- And use `plt.tight_layout()` to adjust spacing betwen subplots

```python
f, axs = plt.subplots(2, 1) #, sharex=True)

axs[0].hist(df['total sulfur dioxide'])
axs[1].hist(df['free sulfur dioxide'])

axs[0].set_xlabel('Total Sulur Dioxide') #<--- notice that we use set_xlabel
axs[1].set_xlabel('Free Sulfur Dioxide') 

axs[0].set_title('Total SO2 Distribution') #<--- notice that we use set_title
axs[1].set_title('Free SO2 Distribution')

axs[0].set_xlim(0, 200)

axs[1].set_yscale('log')

plt.suptitle('SO2 Distributions')

plt.tight_layout()

plt.show()
```

Start practicing with subplots so you can get the hang for them.

If you just want to quickly plot something, you can use the "regular" non-subplot methods to make them quickly.

But if you want to start making more "serious" plots that you can have more control over, use `subplots`

**Pair Plot**

A powerful plot one can use in seaborn is a **pair plot**

This makes a square grid of plots which shows a scatterplot of each feature against itself, as well as a histogram of each feature.

But **be careful** because it can be **very slow** for data sets with many columns (>6) as it makes $N^2$ plots, where $N$ is the number of columns

```python
sns.pairplot(data=iris)#, hue='species')
```

**Correlation Matrix and Heat Maps**

An alternative to using the pair plot, that will still work well with larger dataframes, is to calculate the correlation between each column.

The **pearson correlation coefficient**, $r$, measures the linear relationship between two sets of data. It lies between -1 and +1.

- $r=+1$ means thre is a perfect positive linear relationship between the two given sets of data

- $r=0$ means there is no linear relationship

- $r=-1$ means there is a perfect negative linear relationshp


More info on the **Pearson correlation coefficient**: https://www.scribbr.com/statistics/pearson-correlation-coefficient/

```python
corr_df = iris.corr(numeric_only=True)
corr_df
```

And correlations can be visualized with a **heatmap**

```python
sns.heatmap(corr_df, annot=True) #, cmap='coolwarm')
```
